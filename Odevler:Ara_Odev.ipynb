```python
# En az bir eksik değer içeren satırlardan ilk 20 tanesi
eksik_veri_olan_satirlar = df[df.isnull().any(axis = 1)]
eksik_veri_olan_satirlar.head(20)
```

```python
# indirim_orani ve musteri_puani sütunlarındaki eksik değerleri medyan ile doldurma
df_temizlenmis = df.copy()

df_temizlenmis['indirim_orani'] = df['indirim_orani'].fillna(df_temizlenmis["indirim_orani"].median())
df_temizlenmis['musteri_puani'] = df['musteri_puani'].fillna(df_temizlenmis["musteri_puani"].median())
print(f"doldurmadan önce indirim_oranı: {df['indirim_orani'].isnull().sum()}")
print(f"doldurmadan önce musteri_puani: {df['musteri_puani'].isnull().sum()}")
print(f"Median ile doldurulduktan sonra indirim_puani: {df_temizlenmis['indirim_orani'].isnull().sum()}")
print(f"Median ile doldurulduktan sonra musteri_puani: {df_temizlenmis['musteri_puani'].isnull().sum()}")
```

```python
# odeme_turu ve musteri_tipi sütunlarındaki eksik değerleri mod ile doldurma

df_temizlenmis['odeme_turu'] = df_temizlenmis['odeme_turu'].fillna(df_temizlenmis["odeme_turu"].mode()[0])
df_temizlenmis['musteri_tipi'] = df_temizlenmis['musteri_tipi'].fillna(df_temizlenmis["musteri_tipi"].mode()[0])

print(f"doldurmadan önce odeme_turu: {df['odeme_turu'].isnull().sum()}")
print(f"doldurmadan önce musteri_tipi: {df['musteri_tipi'].isnull().sum()}")
print(f"Mode ile doldurulduktan sonra odeme_turu: {df_temizlenmis['odeme_turu'].isnull().sum()}")
print(f"Mode ile doldurulduktan sonra musteri_tipi: {df_temizlenmis['musteri_tipi'].isnull().sum()}")


```

```python
# siparis_tarihi sütununu datetime tipine dönüştürme
df_temizlenmis["siparis_tarihi"] = pd.to_datetime(df['siparis_tarihi'])

# siparis_yili, siparis_ayi, siparis_gunu ve haftanin_gunu adında 4 yeni sütun üretin
# ilk 10 satırını gösterin

df_temizlenmis["siparis_yili"] = df_temizlenmis["siparis_tarihi"].dt.year
df_temizlenmis["siparis_ayi"] = df_temizlenmis["siparis_tarihi"].dt.month
df_temizlenmis["siparis_gunu"] = df_temizlenmis["siparis_tarihi"].dt.day
df_temizlenmis["haftanin_gunu"] = df_temizlenmis["siparis_tarihi"].dt.weekday

df_temizlenmis.head(10)
```

```python
# Veri setindeki duplicate kayıt sayısı
df_temizlenmis.duplicated().sum()
```

```python
# Duplicate kayıtları temizle
df_temizlenmis = df_temizlenmis.drop_duplicates()
print(f"Veri seti temizlenmeden önce kayıt sayısı: {df.shape[0]}")

print(f"Veri seti temizlendikten sonra kayıt sayısı: {df_temizlenmis.shape[0]}")
```

```python
# sehir sütunundaki benzersiz değerler
df_temizlenmis['sehir'].unique()
```

```python
df_temizlenmis['sehir'] = df_temizlenmis['sehir'].str.replace('I', 'İ').str.title()
df_temizlenmis['sehir'].unique()
# Temizlikten önce ve sonra benzersiz değer sayıları
print(f"Temizlikten önce benzersiz şehir sayısı: {len(df['sehir'].unique())}")
print(f"Temizlikten sonra benzersiz şehir sayısı: {len(df_temizlenmis['sehir'].unique())}")
```

```python
# kategori sütunundaki tutarsız değerleri incele
df_temizlenmis['kategori'].unique()
```

```python
# İçindeki boşlukları temizle ve tamamını baş harfi tek büyük olacak hale getir
df_temizlenmis['kategori'] = df_temizlenmis['kategori'].str.title().str.strip()
df_temizlenmis['kategori'].unique()
```

```python
# Karakter düzenlemeleri
df_temizlenmis['kategori'] = df_temizlenmis['kategori'].str.replace("ı","i").replace("Ev&Yaşam",'Ev Ve Yaşam')
df_temizlenmis['kategori'].value_counts()
```

```python
"""
Mantıksal Hatalar:
   birim_fiyat <= 0,
   toplam_tutar <= 0,
   teslimat_gunu < 0,
   musteri_puani > 5
"""

print(f"birim_fiyat <= 0 : {len(df_temizlenmis[df_temizlenmis['birim_fiyat']<=0])}")
print(f"toplam_tutar <= 0 : {len(df_temizlenmis[df_temizlenmis['toplam_tutar']<=0])}")
print(f"teslimat_gunu <= 0 : {len(df_temizlenmis[df_temizlenmis['teslimat_gunu']<0])}")
print(f"musteri_puani <= 0 : {len(df_temizlenmis[df_temizlenmis['musteri_puani']>5])}")
```

```python
df_temizlenmis = df_temizlenmis[df_temizlenmis['musteri_puani'] <= 5]
```

```python
print(f"musteri_puani <= 0 : {len(df_temizlenmis[df_temizlenmis['musteri_puani']>5])}")
```

```python
# birim_fiyat için IQR
Q1 = df_temizlenmis['birim_fiyat'].quantile(0.25)
Q3 = df_temizlenmis['birim_fiyat'].quantile(0.75)

IQR = Q3 - Q1

alt_sinir = Q1 - 1.5 * IQR
ust_sinir = Q3 + 1.5 * IQR

aykiri_degerler = df_temizlenmis[(df_temizlenmis['birim_fiyat'] < alt_sinir )| (df_temizlenmis['birim_fiyat'] > ust_sinir)]

len(aykiri_degerler)
```
